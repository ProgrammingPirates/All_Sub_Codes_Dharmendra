```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('page_block.csv')

# Separate the class labels
X = df.drop(columns=['Class'])
y = df['Class']

# Compute the means and standard deviations for each class
means = X.groupby(y).mean()
stds = X.groupby(y).std()

# Compute Fisher's score for each attribute
fisher_scores = np.abs(means.iloc[0, :] - means.iloc[1, :]) / (stds.iloc[0, :] + stds.iloc[1, :])

# Assign rank to each attribute
ranked_features = fisher_scores.rank(ascending=False)

# Select top k ranked features
k = 5
selected_features = ranked_features[ranked_features <= k].index.tolist()

# Print the selected features
print(selected_features)
```
