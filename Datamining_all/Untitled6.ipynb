```python
import pandas as pd
import numpy as np
df = pd.read_csv('buys_computer.csv')
def count_labels(attr):
    return df.groupby([attr, 'buys_computer']).size().to_dict()
print("a. Count of labels for each attribute value: ")
print("Age: ", count_labels('age'))
print("Income: ", count_labels('income'))
print("Student: ", count_labels('student'))
print("Credit Rating: ", count_labels('credit_rating'))
def compute_probability(attr):
    prob_df = df.groupby([attr, 'buys_computer']).size().div(len(df)).to_frame(name='probability').reset_index()
    return prob_df
print("\nb. Probability of class labels for each attribute value: ")
print("Age: ", compute_probability('age'))
print("Income: ", compute_probability('income'))
print("Student: ", compute_probability('student'))
print("Credit Rating: ", compute_probability('credit_rating'))
def bayes_theorem(X):
    p_yes = len(df[df['buys_computer'] == 'yes']) / len(df)
    p_no = len(df[df['buys_computer'] == 'no']) / len(df)
    p_X_given_yes = np.prod([compute_probability(attr)[compute_probability(attr)[attr] == X[attr]]['probability'].values[0] for attr in X.keys()]) * p_yes
    p_X_given_no = np.prod([compute_probability(attr)[compute_probability(attr)[attr] == X[attr]]['probability'].values[0] for attr in X.keys()]) * p_no
    p_yes_given_X = p_X_given_yes / (p_X_given_yes + p_X_given_no)
    p_no_given_X = p_X_given_no / (p_X_given_yes + p_X_given_no)

    return {'yes': p_yes_given_X, 'no': p_no_given_X}

X = {'age': 'youth', 'income': 'medium', 'student': 'yes', 'credit_rating': 'fair'}
print("\nc. Probability of instance X belonging to each class label: ")
print(bayes_theorem(X))
# d. Justify the probability of X belongs to a particular class label (buys_computer=”yes” or buys_computer=”no”)
print("\nd. Justification of the probability of X belonging to a particular class label: ")
print("The probability of X belonging to a particular class label is computed using Bayes’ theorem, which provides a way to update our prior beliefs (prior probability) in the light of new evidence (conditional probabilities). The computed probabilities of X belonging to the “yes” and “no” class labels represent the posterior probabilities of the hypothesis given the evidence.")
```
