```python
import pandas as pd
import math

# Load dataset
data = pd.read_csv('buys_computer.csv')

# Compute entropy of the entire dataset
def entropy(dataset):
    n = len(dataset)
    count = dataset['buys_computer'].value_counts()
    entropy = 0
    for value in count.values:
        p = value / n
        entropy += -p * math.log2(p)
    return entropy

# Compute entropy of attribute A
def entropy_attribute(dataset, attribute):
    n = len(dataset)
    values = dataset[attribute].unique()
    entropy = 0
    for value in values:
        subset = dataset[dataset[attribute] == value]
        count = subset['buys_computer'].value_counts()
        p = count.sum() / n
        value_entropy = 0
        for v in count.values:
            p_v = v / count.sum()
            value_entropy += -p_v * math.log2(p_v)
        entropy += p * value_entropy
    return entropy

# Compute information gain of attribute A
def information_gain(dataset, attribute):
    return entropy(dataset) - entropy_attribute(dataset, attribute)

# Select k attributes with highest information gain
def select_k_attributes(dataset, k):
    attributes = dataset.columns[:-1]
    gains = [information_gain(dataset, a) for a in attributes]
    ranked = sorted(zip(gains, attributes), reverse=True)
    return [a for g, a in ranked[:k]]




# Load dataset
data = pd.read_csv('buys_computer.csv')

# Select k attributes with highest information gain
k = 2
selected_attributes = select_k_attributes(data, k)
print(selected_attributes)
```
